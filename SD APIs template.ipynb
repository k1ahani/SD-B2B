## Create profile

```python
url = "https://qa-selfdecode.com/service/profiles/profile/"

payload = {
    "is_v2": True,
    "name": "Test User",
    "ethnicity": "Latino (e.g. Mexican, Peruvian, Colombian)",
    "sex": "Male",
    "birth_year": 1984
}

response = requests.post(url, json=payload, headers=headers)

created_profile = json.loads(response.text)
created_profile
```

## Delete profile

```python
profile_id = created_profile['id']

url = "https://qa-selfdecode.com/service/profiles/profile/{}".format(profile_id)

response = requests.delete(url, headers=headers)
response
```

# PDF generation

```python
url = "https://qa-selfdecode.com/service/pdf-generation-api/reports-pdf-report/"

payload = {
    "white_labelled": False,
    "recommendation_type": "full",
    "report_id": report_id,
    "profile_id": profile_id
}

response = requests.post(url, json=payload, headers=headers)
response
```

```python
request_id = json.loads(response.text)['request_id']
```

```python
url = "https://qa-selfdecode.com/service/pdf-generation-api/reports-pdf-report/?request_id={}".format(request_id)

status = 'pending'
max_iterations = 10
while status != 'success':
    if max_iterations == 0:
        break
    time.sleep(5)
    max_iterations -= 1
    response = requests.get(url, headers=headers)
    status = json.loads(response.text)['status']
    print(".", end="")

pdf_url = json.loads(response.text)['result']
pdf_url
```

```python
wget.download(pdf_url)
```
