## Post individual result

### Numeric markers

```python
def post_numeric_result(profile_id, taken_at, value, marker_id, unit_id):
    url = f"{DOMAIN}/service/labs/numeric-result/"
    
    payload = {"profile_id": profile_id,
               "taken_at": taken_at,
               "value": value,
               "marker_id": marker_id,
               "unit_id": unit_id}
    
    response = requests.post(url,
                             json=payload,
                             headers=headers)
    response.raise_for_status()
    return response.json()
```

## Post results from PDF report

### 1. Get presigned URL to upload lab report file

```python
def labs_presigned_url(filename):
    url = f"{DOMAIN}/service/labs/ocr/get_presigned_url/?file_name={filename}"
    
    response = requests.get(url,
                            headers=headers)
    response.raise_for_status()
    return response.json()["fields"], response.json()["url"]
```

```python
# filename = "CBC Test Results sample.jpg"
filename = "LabCorp sample.pdf"
```

```python
upload_fields, upload_url = labs_presigned_url(filename)
# print(upload_fields)
# print(upload_url)
```

### 2. POST the file to the presigned URL

```python
def upload_labs_file(filename, upload_url, fields):
    with open(filename, "rb") as f:
        files = {"file": (filename, f)}
        response = requests.post(upload_url,
                                 data=fields,
                                 files=files)
        response.raise_for_status()
    print(response)
```

```python
upload_labs_file(filename, upload_url, upload_fields)
```

### 3. Initiate Object Character Recognition (ocr) of the recently uploaded file

```python
def initiate_ocr(key):
    url = f"{DOMAIN}/service/labs/ocr/"
    
    filetypes = {".pdf": "application/pdf",
                 ".png": "image/png",
                 ".jpg": "image/jpg"}
    
    payload = {"file": key,
               "content-type": filetypes[Path(key).suffix]}
    
    response = requests.post(url,
                             json=payload,
                             headers = headers)
    response.raise_for_status()
    return response.json()['request_id']
```

```python
request_id = initiate_ocr(upload_fields['key'])
```

### 4. Query the status of the file

```python
def get_ocr_results(request_id):
    url = f"{DOMAIN}/service/labs/ocr/?request_id={request_id}"
    
    response = requests.get(url,
                            headers = headers)
    response.raise_for_status()
    return response.json()
```

```python
ocr_results = get_ocr_results(request_id)
ocr_results
```

```python
ocr_date = ocr_results['date']
ocr_results_df = pd.json_normalize(ocr_results, 'results')
ocr_results_df
```

### 5. Posting results

```python
profile_id = '4d8b2526-fe5f-4403-a884-8c8f121727db'
```

```python
pd.isna(ocr_results_df.iloc[0]['level'])
```

```python
for index, row in ocr_results_df.iterrows():
    if pd.isna(row['level']):
        print("Item", str(index), ": skipped")
        continue
    else:
        post_numeric_result(profile_id=profile_id,
                            taken_at=ocr_date,
                            value=row['level'],
                            marker_id=row['marker.id'],
                            unit_id=row['unit.id'])
        print("Item", str(index), ": OK")
```
